```python
# Hyper parameters:
# max_df = 0.5
# min_df = 0.05
num_topics = 5
lsi_threshold =0.5
```

```python
tokenized_answers = [answer.split() for answer in answers]
```

```python
import pandas as pd
# from sklearn.feature_extraction.text import CountVectorizer
# from gensim import matutils
# import scipy.sparse
from gensim import corpora
MyDict = corpora.Dictionary(tokenized_answers) 
corpus = [MyDict.doc2bow(text) for text in tokenized_answers]
# cv = CountVectorizer(max_df=max_df, min_df=min_df) 
# Create document term matrix
# data_cv = cv.fit_transform(answers)
# data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
# data_dtm.reset_index()
# tdm = data_dtm.transpose()
# sparse_counts = scipy.sparse.csr_matrix(tdm)
# corpus2 = matutils.Sparse2Corpus(sparse_counts)
# id2word = dict((v, k) for k, v in cv.vocabulary_.items()) 
```

```python
from gensim.models import LsiModel
model = LsiModel(corpus, id2word=MyDict, num_topics=num_topics)
# model2 = LsiModel(corpus2, id2word=id2word, num_topics=num_topics)
```

```python
from gensim.similarities import MatrixSimilarity
lsi_predictions = []
# lsi_predictions2 = []
for i, answer in enumerate(corpus[:l]):
  model_answer = model[corpus[l+i]]
  # model_answer2 = model[corpus2[l+i]]
  index = MatrixSimilarity(model[corpus[i:i+1]], num_features=num_topics)
  # index2 = MatrixSimilarity(model2[corpus2], num_features=num_topics)
  sims = index[model_answer][0]
  # sims2 = index2[model_answer2][0]
  lsi_predictions.append(1 if sims > lsi_threshold else 0)
  # lsi_predictions2.append(1 if sims2 > lsi_threshold else 0)
```

```python
acc =  accuracy_score(data['accuracy'].to_numpy(), lsi_predictions)
# acc2 = accuracy_score(data['accuracy'].to_numpy(), lsi_predictions2)
```
