```python
import nltk
nltk.download('stopwords')
#nltk.download()
import pandas as pd
nltk.download('punkt')
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import preprocessing
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.decomposition import TruncatedSVD


data = pd.read_csv('C:/Users/Eng/Desktop/Dataset.csv')
label_encoder = preprocessing.LabelEncoder()
data ['accuracy'] = label_encoder.fit_transform(data['accuracy'])
student_answer = data['studentAnswer']
ref_answers = data['referenceAnswer']
ps = PorterStemmer()
#answers_arr = answers.to_numpy()
#ex_sentence = answers[0]
stop_words = set(stopwords.words("english"))
```

```python
def clean(s):
  out = word_tokenize(s)
  ans = ' '.join([''.join(ps.stem(word)) for word in out if word not in stop_words])
  return ans
filtered_answers = [clean(s) for s in student_answer]
```

```python
vectorizer = TfidfVectorizer( stop_words=stop_words , use_idf=True , ngram_range=(1,3))
X = vectorizer.fit_transform(filtered_answers)
```

```python
lsa = TruncatedSVD(n_components=27 , n_iter=100)
lsa.fit(X)
```
